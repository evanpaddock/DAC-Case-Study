```python
# Import necessary dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in categories data
categories_df = pd.read_csv("./data/amazon_categories.csv", index_col="id")
categories_df.head()
```

```python
# Find category 13
categories_df.iloc[12]
```

```python
# Read in products data
products_df = pd.read_csv("./data/amazon_products.csv", index_col="asin")
products_df.head()
```

```python
# Find the product with 346563 reviews
products_df.loc[products_df["reviews"] == 346563]
```

```python
# Q1. Show 10 products that are best sellers.
products_df[products_df["isBestSeller"] == True].head(10)
```

```python
# Q2. Find the top 25 best selling products by # sold in last month.
asin_numSold_top20 = products_df["boughtInLastMonth"]

asin_numSold_top20 = asin_numSold_top20.nlargest(25)
# asin_numSold_top20 = asin_numSold_top20.sort_values(ascending=False).head(25)

asin_numSold_top20.plot(kind='bar', color='skyblue')
```

```python
# Q3. Merge products and categories into a singular dataframe
cat_prod_df = pd.merge(categories_df, products_df, "inner", left_index=True, right_on="category_id")
cat_prod_df.head()
```

```python
# Q4. Find the top 3 best selling categories by # sold last month
top5_best_selling_by_cat = cat_prod_df.groupby("category_name")["boughtInLastMonth"].sum()
top5_best_selling_by_cat.sort_values(ascending=False).head(3)
# top5_best_selling_by_cat.nlargest(3)
```

```python
# Q5. Which categories did not sell a single item last month
df = cat_prod_df.reset_index()
# df.groupby("category_name")["boughtInLastMonth"].min()

aggregated_df = df.groupby('category_name').agg(
    total_price=('price', 'sum'),
    total_bought=('boughtInLastMonth', 'sum')
)

aggregated_df[aggregated_df["total_bought"] == 0]
```
